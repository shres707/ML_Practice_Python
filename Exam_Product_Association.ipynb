```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

from mlxtend.frequent_patterns import apriori

from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_csv('C:/Users/G V Swetha/Desktop/PA.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.drop(['pmethod','homeown','sex'],axis=1,inplace=True)
```

```python
df.shape
df
```

```python
df1=pd.get_dummies(df)
df1.head()
```

```python
frequent_items=apriori(df1,min_support=0.05,use_colnames=True)
frequent_items.sort_values(by='support',ascending=False)
```

```python
rules=association_rules(frequent_items,metric='lift',min_threshold=1)
```

```python
type(rules)
```

```python
rules.sort_values(by='lift',ascending=True)
```

```python
rules.describe()
```

```python
rules[(rules['lift']>=1) & (rules['confidence']>=0.8) & (rules['support']>=0.2)]
```
