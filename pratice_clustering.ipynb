```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_min=df[['gmp','gi','cogs','rating','quantity','unit_price']]
```

```python
df_min
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
scld=MinMaxScaler(feature_range=(0,1))
arr_scld=scld.fit_transform(df_min)
df_scld=pd.DataFrame(arr_scld,columns=df_min.columns)
df_scld.head()
df_scld.describe()
df_scld.shape
import warnings
warnings.filterwarnings('ignore')
df_min.describe()
```

```python
from sklearn.cluster import KMeans
num_of_clusters = range(2,10)
error=[]
for num_clusters in num_of_clusters:
    clusters=KMeans(num_clusters)
    clusters.fit(df_scld)
    error.append(clusters.inertia_/100)
```

```python
df=pd.DataFrame({"cluster_Numbers":num_of_clusters,"Error_Term":error})
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot(df.cluster_Numbers,df.Error_Term,marker="D",color="#a51347")
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.title("Find the optimal number of Cluster")
plt.show()
          
```

```python
clusters=KMeans(3)
clusters.fit(df_scld)
clusters.labels_
```

```python
df_min['Cluster']=clusters.labels_
```
